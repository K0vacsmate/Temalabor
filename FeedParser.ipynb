```python
import feedparser as fp
import pandas as pd
from dateutil.parser import parse
import numpy as np
```

```python
companies={
    "Microsoft":["Microsoft", "Windows", "XBox"],
    "Apple":["Apple", "IOS", "IPhone"],
    "Amazon":["Amazon", "AWS"],
    "Alphabet": ["Alphabet", "Google", "Android"],
    "Facebook":["Facebook", "Zuckerberg"],
    "Intel":["Intel", "i3", "i5", "i7", "i9"],
    "Adobe":["Adobe", "Photoshop", "Lightroom"],
    "Verizon":["Verizon"],
    "NVIDIA":["NVIDIA", "RTX", "GeForce"],
    "Netflix":["Netflix"],
    "AT&T":["AT&T"],
    "AMD":["AMD", "Zen", "Zen2", "Ryzen", "Vega"],
    "DELL":["DELL"],
    "Lenovo":["Lenovo"],
    "Qualcomm":["Qualcomm", "Snapdragon"]
}
```

```python
url="https://www.cnbc.com/id/19854910/device/rss/rss.html"
f=fp.parse(url)
```

```python
news=[]
for i in range(len(f.entries)):
    news.append(tuple([f.entries[i].title, parse(f.entries[i].published).date(), f.entries[i].link]))
    
newsDf=pd.DataFrame(news)
newsDf.columns=['Title', 'Date', 'Link']
newsDf=newsDf.sort_values(by=['Date'])
newsDf
```

```python
newsComp=[]

for brand in companies:
    for name in companies[brand]:
        for title in newsDf['Title']:
            if (name in title):
               newsComp.append(tuple([brand, title]))
            
newsComp=pd.DataFrame(newsComp)
newsComp.columns=['Company','Title']
N=newsComp.join(newsDf.set_index('Title'), on='Title')
N['Date']=N['Date'].astype(str)
N['Company']=N['Company'].astype(str)
N
```

```python
diffs=pd.read_csv('diffs.csv')
diffs.set_index('Date', inplace=True)
diffs=diffs.stack()
diffs=pd.DataFrame(diffs)
diffs=diffs.reset_index()
diffs.columns=['Date','Company','Change']
diffs['Date']=diffs['Date'].astype(str)
diffs['Company']=diffs['Company'].astype(str)
diffs
```

```python
final=pd.merge(diffs, N, how='right',left_on=['Date', 'Company'], right_on=['Date', 'Company'])
final.groupby('Date')
final
```

```python
final.to_csv('Final.csv')
```
