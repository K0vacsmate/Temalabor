```python
import yfinance as yf
import pandas as pd
import datetime as dt
```

```python
companies={
    "MSFT":"Microsoft",
    "AAPL":"Apple",
    "AMZN":"Amazon",
    "GOOGL":"Alphabet",
    "FB":"Facebook",
    "INTC":"Intel",
    "ADBE":"Adobe",
    "VZ":"Verizon",
    "NVDA":"NVIDIA",
    "NFLX":"Netflix",
    "T":"AT&T",
    "AMD":"AMD",
    "DELL":"DELL",
    "LNVGY":"Lenovo",
    "QCOM":"Qualcomm"
}
```

```python
today=dt.date.today()
symbols=list(companies.keys())
tickers=pd.DataFrame(yf.download(symbols, start="2020-05-01", end=today))
tickers
```

```python
OpenClose=pd.DataFrame(tickers[['Open','Close']])
OpenClose
```

```python
Gains=pd.DataFrame(round(OpenClose['Close']/OpenClose['Open']*100-100,2))
mean=round(Gains.mean(axis=1),2)
Gains=Gains.rename(columns=companies)
Gains['mean']=mean
Gains
```

```python
Diffs=Gains.sub(mean.values, axis=0)
Diffs.drop(['mean'],axis=1)
Diffs
```

```python
Gains.to_csv('gains.csv')
Diffs.to_csv('diffs.csv')
```
